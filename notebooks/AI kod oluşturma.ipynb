````python
import google.generativeai as genai
import pandas as pd
import time
import random
import os

# ==========================================
# AYARLAR
# ==========================================
API_KEY = "REMOVEDDNvXH34Iafd-FrxdgeArs1JC8gYJ6FDBQ"
FILE_NAME = "python_ai_dataset_7000.xlsx"
TARGET_COUNT = 7000

# Gemini Yapılandırması
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash')

topics = [
    "pandas dataframe filtering", "numpy array multiplication", "matplotlib plotting line chart",
    "seaborn heatmap", "reading excel with pandas", "handling missing values pandas",
    "flask simple route", "django model definition", "requests get library", 
    "beautifulsoup web scraping", "selenium browser automation", "fastapi async endpoint",
    "os module file listing", "shutil move file", "reading writing text file",
    "subprocess run command", "json dump and load", "csv writer",
    "list comprehension example", "dictionary merge", "lambda function usage",
    "generator yield example", "decorator implementation", "class inheritance",
    "binary search implementation", "sorting a list of dictionaries",
    "regex phone number match", "datetime current time", "random password generator",
    "logging to file", "try except error handling", "multiprocessing pool",
    "asyncio basic example", "sqlite3 connection", "sending email with smtplib"
]

styles = [
    "beginner level", "production ready", "optimized one-liner", 
    "object oriented", "functional style", "buggy code for testing"
]

# ==========================================
# SADELEŞTİRİLMİŞ DÖNGÜ
# ==========================================

if os.path.exists(FILE_NAME):
    try:
        df_existing = pd.read_excel(FILE_NAME)
        data_list = df_existing.to_dict('records')
        current_count = len(data_list)
    except:
        data_list = []
        current_count = 0
else:
    data_list = []
    current_count = 0

consecutive_errors = 0 

print(f"Başlanıyor... Hedef: {TARGET_COUNT}")
print(f"Mevcut Sayı: {current_count}")

while current_count < TARGET_COUNT:
    try:
        topic = random.choice(topics)
        style = random.choice(styles)
        
        prompt = (
            f"Write a Python code snippet about {topic}. Style: {style}. "
            "Output ONLY raw Python code. No markdown, no comments."
        )

        response = model.generate_content(prompt)
        clean_code = response.text.replace("```python", "").replace("```", "").strip()
        
        data_list.append({
            "id": current_count + 1,
            "code": clean_code,
            "language": "Python",
            "topic": topic,
            "label": 1,
            "source": "Gemini-2.0-Flash"
        })
        
        current_count += 1
        consecutive_errors = 0 
        
        # SADECE SAYI YAZDIRILIYOR
        print(f"Toplanan Kod Sayısı: {current_count}")

        # Her 20'de bir kaydet (Sessizce)
        if current_count % 20 == 0:
            df = pd.DataFrame(data_list)
            df.to_excel(FILE_NAME, index=False)
            
        time.sleep(10) 

    except Exception:
        consecutive_errors += 1
        # Hata mesajı basmıyoruz, sadece bekleme durumunu belirtiyoruz
        wait_time = min(consecutive_errors * 60, 3600)
        print(f"...Kota/Hata oluştu. {wait_time} saniye bekleniyor (Sayı: {current_count})...")
        time.sleep(wait_time)

# Bitiş
df = pd.DataFrame(data_list)
df.to_excel(FILE_NAME, index=False)
print("Tamamlandı.")
````
